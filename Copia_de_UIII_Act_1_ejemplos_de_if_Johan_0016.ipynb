```python
#4: Sentencia if-elif-else integrando todos los operadores
#Este código categoriza el desempeño de un estudiante basándose en un puntaje, aplicando un bono si el examen fue extra largo.
puntaje_base = 75  # Puntaje obtenido por el estudiante [11].
bono_extra = 10  # Puntos adicionales por dificultad [5].
total = puntaje_base + bono_extra  # Operador aritmético '+' para calcular el puntaje final [5].

if total >= 90:  # Verifica si el total es mayor o igual a 90 [8].
    print("Calificación: Excelente")
elif total >= 70 and total < 90:  # 'and' combina dos comparaciones relacionales [8, 10].
    print("Calificación: Aceptable")  # Se ejecuta si el total está en el rango [70, 90) [14].
elif total >= 60:  # Evalúa esta opción solo si las anteriores fueron falsas [1].
    print("Calificación: Suficiente")
else:
    print("Calificación: Insuficiente")  # Acción por defecto si nada anterior se cumplió [1].
```
